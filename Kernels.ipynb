```python
# put the local directory files/python in path
import sys
sys.path.append('files/python')
import pylab as plt
import numpy as np
```

This notebook explains and demonstrates the basic use of the code in [`files/python/kernels.py`](files/python/kernels.py).

The main code in there is the class `Kernels`. It is an implementation of a variety of linear kernel-driven BRDF models.

```python
# how to import the class

import kernels
```

The file contains a test function `mimic` that should reproduce the results in Wanner et al. 1995.

This is a good way to test that the code is operating as intended.

```python
result = kernels.mimic(doPrint=False,doPlot=True,RossHS=False,RecipFlag=False)
```

That said, you should note that soon after then, the modifications were made to the kernels to enable reciprocity, and these are the ones we tend to use:

```python
result = kernels.mimic(doPrint=False,doPlot=True,RossHS=False,RecipFlag=True)
```

Running these for some other set of angles then, we set up arrays of solar zenith angle, view zenith angle and relative azimuth angle (degrees) (`sza, vza, raa`), e.g.:

```python
vza = np.arange(-70,70,1.)
sza = np.zeros_like(vza) + 45.
raa = np.zeros_like(vza)
```

and run the kernels code, e.g. for RossThick, LiSparseReciprocal (the 'MODIS kernels'):

```python
kk = kernels.Kernels(vza,sza,raa,\
                     RossHS=False,MODISSPARSE=True,\
                     RecipFlag=True,normalise=1,\
                     doIntegrals=False,LiType='Dense',RossType='Thick')
```

The kernel values then are contained within:

```python
ross = kk.Ross
li   = kk.Li
iso =  kk.Isotropic
```

```python
plt.plot(vza,ross)
plt.title('RossThick sza = 45')
plt.xlabel('view zenith angle')
plt.ylabel('kernel value')
```

```python
plt.plot(vza,li)
plt.title('LiSparse sza = 45')
plt.xlabel('view zenith angle')
plt.ylabel('kernel value')
```

Form the Li kernel, we can see that the hot spot is when `raa=0` and `vza = sza = 45.`.

By default, kernels (other than the Isotropic) are defined to be zero at `vza = sza = 0`:

```python
kk0 = kernels.Kernels(vza,sza * 0.0,raa,\
                     RossHS=False,MODISSPARSE=True,\
                     RecipFlag=True,normalise=1,\
                     doIntegrals=False,LiType='Dense',RossType='Thick')

print kk0.Ross[vza == 0],kk0.Li[vza == 0]
```
